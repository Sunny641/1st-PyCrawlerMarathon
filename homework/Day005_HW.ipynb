```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data = json.loads(r.text)

print(len(data))
print(data[0].keys())
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for d in data:
    print(d['title'])
    print(d['createdAt'])
    print(d['commentCount'])
    print(d['likeCount'])
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
commentCount = 0
likeCount = 0

for d in data:
    commentCount += d['commentCount']
    likeCount += d['likeCount']

print('ç†±é–€æ–‡ç« ')    
print('å¹³å‡ç•™è¨€äººæ•¸: ',commentCount / len(data)) 
print('å¹³å‡æŒ‰è®šäººæ•¸: ',likeCount / len(data)) 

r2 = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
data2 = json.loads(r2.text)
commentCount = 0
likeCount = 0

for d in data2:
    commentCount += d['commentCount']
    likeCount += d['likeCount']
    
print('éç†±é–€æ–‡ç« ')    
print('å¹³å‡ç•™è¨€äººæ•¸: ',commentCount / len(data)) 
print('å¹³å‡æŒ‰è®šäººæ•¸: ',likeCount / len(data)) 
```
