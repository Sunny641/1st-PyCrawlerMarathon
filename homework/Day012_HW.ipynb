### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

i = 0
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('em')[0].text,  d.find_all('a')[-1].text)
    i = i + 1
    
print()
print('總共文章數: %d' %i)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
dict_1 = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    category = d.find_all('em')[0].text
    #print(category)
    
    if (dict_1.get(category,'N') == 'N'):
        dict_1[category] = 1
    else:
        dict_1[category] = dict_1[category] + 1
    
print(dict_1.items())
print()
print('類別有 {0} 種'.format(len(dict_1)))
```

### ③ 哪一個類別的文章最多

```python
#最多文章的文章數
maxnum = max(dict_1.values())

for k,v in dict_1.items():
    if v == maxnum:
        #print('{0}類有{1}篇文章'.format(k,v))
        print('最多文章數(',maxnum,')的類別是:',k,end=' ')
```
