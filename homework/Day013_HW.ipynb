## 作業目標

根據範例 ，完成以下問題：

* ① 印出最新文章的「作者」「標題」「時間」
* ② 印出第一頁所有文章的「作者」「標題」「時間」


### ① 印出最新文章的「作者」「標題」「時間」

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/NBA/index.html'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

d = soup.find_all(class_="r-ent")[0]
print('最新文章:')
print('\t作者:',d.find(class_="author").text.replace('\t', '').replace('\n', ''))
print('\t標題:',d.find(class_="title").text.replace('\t', '').replace('\n', ''))
print('\t時間:',d.find(class_="date").text.replace('\t', '').replace('\n', ''))
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
for index, d in enumerate(soup.find_all(class_="r-ent"), start = 1):
    print('文章 %d' %index)
    print('\t作者:',d.find(class_="author").text.replace('\t', '').replace('\n', ''))
    print('\t標題:',d.find(class_="title").text.replace('\t', '').replace('\n', ''))
    print('\t時間:',d.find(class_="date").text.replace('\t', '').replace('\n', ''))
    
```

### ③ 試著爬爬看其他版的文章

```python
#美食版
url = 'https://www.ptt.cc/bbs/Food/index.html'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html5lib")

for index, d in enumerate(soup.find_all(class_="r-ent"), start = 1):
    print('文章 %d' %index)
    print('\t作者:',d.find(class_="author").text.replace('\t', '').replace('\n', ''))
    print('\t標題:',d.find(class_="title").text.replace('\t', '').replace('\n', ''))
    print('\t時間:',d.find(class_="date").text.replace('\t', '').replace('\n', ''))
    
```
